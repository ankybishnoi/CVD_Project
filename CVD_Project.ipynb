```python
x = df.drop(['target','fbs','chol','restecg','age'],axis=1)
y = df['target']
    
```

```python
x = sm.add_constant(x)
```

```python
model = sm.OLS(y,x)
rezult = model.fit()
print(rezult.summary())
```

```python
from sklearn.model_selection import train_test_split

train_x,test_x, train_y,  test_y = train_test_split(x,y,test_size =0.20, random_state=129)
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
from sklearn.linear_model import LogisticRegression

Lr = LogisticRegression()
```

```python
Lr.fit(train_x,train_y)
```

```python
predicted = pd.DataFrame(Lr.predict(test_x))
```

```python
from sklearn import metrics

print(metrics.accuracy_score(test_y,predicted))
print(metrics.classification_report(test_y,predicted))
```

```python
from sklearn.ensemble import RandomForestClassifier

rm = RandomForestClassifier(n_estimators=100, max_depth=4, random_state=129)

rm.fit(train_x, train_y)
```

```python
predicted = pd.DataFrame(rm.predict(test_x))

print(metrics.accuracy_score(test_y,predicted))
print(metrics.classification_report(test_y,predicted))
print(metrics.confusion_matrix(test_y,predicted))

```
